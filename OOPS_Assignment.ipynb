```python
#(1). What are the five key concepts of Object-Oriented Programming (OOP)?

# Here are five key concept of object-oriented programming

# object oriented programming(oop):-
# object oriented prgramming is a programming paradigm that to design application and computyers program
#Oop is based on several concept such as 
#(i). Class Object 
#(ii). Encapsulation
# (iii). Abstraction,
#(iv). Inheritance
#(v). polymorphism
# it aim to imitate and simplify the real world by programmjng object that cantain both data and function

#(i). Class >> Class is user define function blue print or protocol that describe the structure and behavior of an object,
# it is data type that holds its own data member and member function.

# object >> An objec is instance of a class It is a self-contained entity with its own set of data and code object can interating
# with each each other through method
# Example >>
class Student:
    def __init__(self,name,course):
        self.name = name     # data member 
        self.course = course
    def std_details(self): # Method / function member
        return f"Student name: {self.name} and Enrolled course {self.course}"

# making the object
std1 = Student("Ashif iqbal","MCA")    # std1 is the class object
print(std1.std_details())

# (ii). Encapsulation>>
# Encapsulation is the binding of a data and method onto a single unite called an object it is used to hide the internal
# implementation details of an object form the outside world, This make the code more sevure and resable

#Example >>
class Employee:
    def __init__(self,name,post,salary):
        self.name = name
        self.post = post
        self.salary = salary
    def emp_details(self):
        return f"name of employee {self.name} Designation of employee: {self.post}"

    def salary_details(self):
        return f"Salary of Emp: {self.salary}"

# public Abstraction

emp1 = Employee("Abdul","Data Engineer",50000)
print(emp1.emp_details())
print(emp1.salary_details())

#(iii). Abstraction:-
# Abstraction is abstraction is the act of hididng of unneccessary details and showing only essential information in oop.
#Abtraction is used to create classes that represent real-world entities. These classes hide the internal implementation
# details of the entities and expose only the necessary information to the outside world

class Bank:
    def __init__(self,name,account):
        self.name = name
        self.account = account
    def customer_details(self):
        return f"Customer name: {self.name} and Account no. {self.account}"

cust = Bank("Ashif",123)
print(cust.customer_details())

# private>> private means that can be accessible only the base class and derived class

class Bank:
    def __init__(self,name,account):
        self._name = name
        self._account = account
    def customer_details(self):
        return f"Customer name: {self._name} and Account no. {self._account}"

cust = Bank("Abdul",123)
print(cust.customer_details())

# Protected >> The protected means that can be accessible only the base class
class Bank:
    def __init__(self,name,account):
        self.__name = name
        self._account = account
    def customer_details(self):
        return f"Customer name: {self.__name} and Account no. {self._account}"

cust = Bank("Iqbal",123)
print(cust.customer_details())

# (iv). Inheritance:-
# Inheritance is ability of a class to inherite the the propertises and methods of anbother class, This allows 
# us to create nre classes that are based on existing class,
# Inheritance makes code reuse and help to reduce code duplication

# Example:-
# 1. Single Inheritance
# In single inheritance, a class (child) inherits from only one base class (parent).

# Parent Class
class Animal:
    def speak(self):
        print("Animal speaks")

# Child Class
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Object of the child class
dog = Dog()
dog.speak()  # Inherited from Animal class
dog.bark()   # Method of Dog class


# 2. Multiple Inheritance
# In multiple inheritance, a class inherits from more than one base class.

# Parent Class 1
class Animal:
    def speak(self):
        print("Animal speaks")

# Parent Class 2
class Mammal:
    def has_fur(self):
        print("Mammal has fur")

# Child Class
class Dog(Animal, Mammal):
    def bark(self):
        print("Dog barks")

# Object of the child class
dog = Dog()
dog.speak()  
dog.has_fur() 
dog.bark()  # Method of Dog


# 3. Multilevel Inheritance
# In multilevel inheritance, a class is derived from a class that is already derived from another class.

# Grandparent Class
class Animal:
    def speak(self):
        print("Animal speaks")

# Parent Class
class Mammal(Animal):
    def has_fur(self):
        print("Mammal has fur")

# Child Class
class Dog(Mammal):
    def bark(self):
        print("Dog barks")

# Object of the child class
dog = Dog()
dog.speak()  # Inherited from Animal
dog.has_fur()  # Inherited from Mammal
dog.bark()  # Method of Dog


# 4. Hierarchical Inheritance
# In hierarchical inheritance, multiple classes inherit from the same base class.

# Parent Class
class Animal:
    def speak(self):
        print("Animal speaks")

# Child Class 1
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Child Class 2
class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Objects of the child classes
dog = Dog()
cat = Cat()
dog.speak()  # Inherited from Animal
dog.bark()   # Method of Dog
cat.speak()  # Inherited from Animal
cat.meow()   # Method of Cat


# 5. Diamond (Hybrid) Inheritance
# Diamond inheritance is a combination of two or more types of inheritance. It often leads to the "
# diamond problem" when the same base class is inherited through multiple paths.


# Base Class
class Animal:
    def speak(self):
        print("Animal speaks")

# Intermediate Class 1
class Mammal(Animal):
    def speak(self):
        print("Mammal speaks")

# Intermediate Class 2
class Bird(Animal):
    def speak(self):
        print("Bird speaks")

# Derived Class (from both Mammal and Bird)
class Bat(Mammal, Bird):
    pass

# Object of the derived class
bat = Bat()
bat.speak()  # Which speak() method will be called? Mammal or Bird?


# 6. Multi-valued Inheritance
# Multi-valued inheritance refers to a situation where multiple child classes inherit multiple attributes or
# methods from a base class.


# Parent Class with multiple methods
class Animal:
    def walk(self):
        print("Animal walks")
        
    def eat(self):
        print("Animal eats")

# Child Classes
class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Objects of child classes
dog = Dog()
cat = Cat()
dog.walk()  # Inherited from Animal
dog.eat()   # Inherited from Animal
dog.bark()  # Dog's method
cat.walk()  # Inherited from Animal
cat.eat()   # Inherited from Animal
cat.meow()  # Cat's method

# (v). Polymorphism:- 
# Polymorphism is the ability of onject of different classes to respond to the same message in different
# ways This allow us to write code that more flexikble and reusable.

# 1. Method Overriding (Run-time Polymorphism)
# In Python, you can define methods in a subclass with the same name as a method in the superclass. This is called method overriding. The method that gets called depends on the type of object that is calling it, enabling run-time polymorphism.

# Example of Method Overriding:

# Base Class
class Animal:
    def sound(self):
        print("This is a generic animal sound")

# Derived Class 1
class Dog(Animal):
    def sound(self):
        print("Dog barks")

# Derived Class 2
class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Polymorphism in action
def make_sound(animal):
    animal.sound() 

# Objects of different classes
dog = Dog()
cat = Cat()

make_sound(dog)  # Output: Dog barks
make_sound(cat)  # Output: Cat meows

# 2. Method Overloading (Compile-time Polymorphism)
# Python doesn’t support method overloading like other languages such as Java or C++. However, you can achieve similar functionality by using default parameters in methods or handling multiple argument types inside a single method.

class Calculator:
    def add(self, a, b, c=0):  # c has a default value of 0
        return a + b + c

# Creating object
calc = Calculator()

# Method Overloading Effect
print(calc.add(5, 10))    # Output: 15 (calls add with 2 arguments)
print(calc.add(5, 10, 15))  # Output: 30 (calls add with 3 arguments)



# 3. Operator Overloading
# Python allows operator overloading, where operators like +, -, *, etc.,
# can be given a different meaning depending on the operands. This is a form of polymorphism where
# the behavior of operators changes based on the types of operands.

# Example of Operator Overloading:
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the '+' operator
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Point({self.x}, {self.y})"

# Creating objects
p1 = Point(1, 2)
p2 = Point(3, 4)

# Using overloaded + operator
result = p1 + p2
print(result)  # Output: Point(4, 6)

# 4. Duck Typing
# Python follows the concept of duck typing, which is a dynamic form of polymorphism. In 
# duck typing, the actual class of the object is less important than the methods or attributes 
# the object has. If an object behaves like a duck (i.e., has methods similar to a duck), it’s treated as a duck.


class Bird:
    def fly(self):
        print("Bird is flying")

class Airplane:
    def fly(self):
        print("Airplane is flying")

# Function that takes any object with a fly method
def fly_in_the_sky(flying_object):
    flying_object.fly()  # Calls the fly method of the object

# Objects of different classes
bird = Bird()
airplane = Airplane()

# Duck typing in action
fly_in_the_sky(bird)      # Output: Bird is flying
fly_in_the_sky(airplane)  # Output: Airplane is flying


# 5. Polymorphism with Functions and Objects
# Polymorphism can also be seen when we create functions that can take objects from 
# different classes as arguments. These objects should share a common interface (method) for polymorphism to work.

# Example:

class Dog:
    def make_sound(self):
        print("Dog barks")

class Cat:
    def make_sound(self):
        print("Cat meows")

class Cow:
    def make_sound(self):
        print("Cow moos")

# Polymorphic function
def animal_sound(animal):
    animal.make_sound()

# Calling the function with different objects
dog = Dog()
cat = Cat()
cow = Cow()

animal_sound(dog)  # Output: Dog barks
animal_sound(cat)  # Output: Cat meows
animal_sound(cow)  # Output: Cow moos

```

```python
# 3. Explain the difference between instance methods and class methods. Provide an example of each.
# Instance Methods vs. Class Methods
# Instance Methods:

# These methods are associated with an instance of a class. They take self as the first parameter, which refers to the specific instance (object) that calls the method.
# Instance methods can access and modify the attributes of the object.
# Example:

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    # Instance method
    def display_info(self):
        return f"Car Info: {self.year} {self.make} {self.model}"

car1 = Car("Honda", "Civic", 2022)
print(car1.display_info())  # Calls the instance method on car1

Class Methods:

# These methods are associated with the class itself rather than an instance. They take cls as the first parameter, which refers to the class (not the object).
# Class methods are used when you want to access or modify class-level attributes.
# You use the @classmethod decorator to define a class method.
# Example:


class Car:
    car_count = 0  # Class attribute

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        Car.car_count += 1

    # Class method
    @classmethod
    def total_cars(cls):
        return f"Total cars created: {cls.car_count}"

car1 = Car("Honda", "Civic", 2022)
car2 = Car("Toyota", "Corolla", 2021)
print(Car.total_cars())  # Calls the class method on the Car class
```

```python
# 5. What are the three types of access modifiers in Python? How are they denoted?

# Python, there are three types of access modifiers that control the visibility and accessibility of 
# class attributes and methods. Python does not have strict access control like some other languages 
# (e.g., private, protected, and public in Java or C++), but it uses naming conventions to indicate the 
# intended access level.

# 1. Public:
# Denoted by: No underscore prefix (normal naming)
# Description: Public members (attributes or methods) are accessible
# from anywhere, both inside and outside the class. By default, all class members are public in Python.
# Example:


class Car:
    def __init__(self, make, model):
        self.make = make  
        self.model = model  # Public attribute

    def display_info(self):  # Public method
        return f"{self.make} {self.model}"

car = Car("Toyota", "Camry")
print(car.make)  # Accessible from outside
print(car.display_info())  


# 2. Protected:
# Denoted by: A single underscore _ prefix
# Description: Protected members are intended to be accessed within the class and its subclasses.
# They are not enforced as private but serve as a convention to indicate that they should not be accessed
# directly outside the class. Python treats them as public but encourages developers to avoid accessing them directly.
# Example:

class Car:
    def __init__(self, make, model):
        self._make = make  
        self._model = model  

    def _display_info(self):  
        return f"{self._make} {self._model}"

car = Car("Toyota", "Camry")
print(car._make)  
print(car._display_info())  


# 3. Private:
# Denoted by: A double underscore __ prefix
# Description: Private members are meant to be inaccessible from outside the class. Python 
# uses name mangling to make private members harder to access, but they can still be accessed indirectly.
# This prevents direct access to such attributes or methods from outside the class.
# Example:


class Car:
    def __init__(self, make, model):
        self.__make = make  
        self.__model = model 
        
    def __display_info(self):  
        return f"{self.__make} {self.__model}"

    def get_info(self): 
        return self.__display_info()

car = Car("Toyota", "Camry")
# print(car.__make)  
print(car.get_info())  # Accessing private members through a public method
```

```python
# 7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

# How does MRO Works:
# In single inheritance, MRO is straightforward, moving from the child class up to the parent.
# In multiple inheritance, MRO ensures that each parent class is checked only once, respecting the order of inheritance (from left to right) and ensuring that a subclass comes before its parent class.
# Example: MRO in Single Inheritance

class A:
    def process(self):
        print("Method in A")

class B(A):
    def process(self):
        print("Method in B")

b = B()
b.process()  # Output: Method in B


# Example: MRO in Multiple Inheritance

class A:
    def process(self):
        print("Method in A")

class B(A):
    def process(self):
        print("Method in B")

class C(A):
    def process(self):
        print("Method in C")

class D(B, C):  # Multiple inheritance
    pass

d = D()
d.process()  


# How to Retrieve MRO Programmatically
# You can retrieve the MRO of a class using:

# The mro() method
# The __mro__ attribute
# Example using mro():


class A: pass
class B(A): pass
class C(A): pass
class D(B, C): pass

# Retrieve MRO
print(D.mro())


# Example using __mro__ attribute:

print(D.__mro__)

```

```python
# 8. Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses
# Circle and Rectangle that implement the area() method.
# To create an abstract base class Shape with an abstract method area(), we can use the abc module in Python,
# which allows us to define abstract base classes. Subclasses like Circle and Rectangle will implement the area() method.

# Here can implement this:


from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass  # Abstract method, must be implemented by subclasses

# Subclass Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    
    def area(self):
        return math.pi * self.radius ** 2

# Subclass Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

  
    def area(self):
        return self.width * self.height

# Example usage:
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle area: {circle.area():.2f}")  
print(f"Rectangle area: {rectangle.area()}")  
```

```python
# 10. Implement encapsulation in a `BankAccount` class with private attributes for `balance` and
# `account_number`. Include methods for deposit, withdrawal, and balance inquiry.

# Encapsulation restricts access to certain class attributes and methods. Here's an implementation using private attributes:


class BankAccount:
    def __init__(self, account_number, balance=0):
        self.__account_number = account_number  
        self.__balance = balance 

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}")
        else:
            print("Insufficient funds or invalid withdrawal amount")

    def get_balance(self):
        return self.__balance

# Example usage
account = BankAccount("123456")
account.deposit(5000)
account.withdraw(500)
print(f"Current balance: {account.get_balance()}")
```

```python
# 11. Write a class that overrides the `__str__` and `__add__` magic methods. What will these methods allow
# you to do?

# Magic methods like __str__ and __add__ are special methods that allow customization of class behavior in Python.

# __str__: Allows you to define what gets printed when an object is passed to print().
# __add__: Allows you to define the behavior of the + operator for your class objects.


class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overriding __str__ to provide a
    def __str__(self):
        return f"Point({self.x}, {self.y})"

    # Overriding __add__ to allow addition of two points
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

# Example usage
p1 = Point(2, 3)
p2 = Point(4, 5)
p3 = p1 + p2  # Uses __add__

print(p3)  
```

```python
# 12. Create a decorator that measures and prints the execution time of a function.
# A decorator can be used to measure and print the execution time of a function:


import time

def measure_time(func):
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
       
    return wrapper

@measure_time
def example_function():
    print(100*254)

# Example usage
print(example_function()) 
```

```python
# 13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

# The Diamond Problem occurs in multiple inheritance when a class inherits from two classes
# that both inherit from the same parent class, causing ambiguity about which method should be called from the parent class.

# Example:


class A:
    def show(self):
        print("Method from A")

class B(A):
    def show(self):
        print("Method from B")

class C(A):
    def show(self):
        print("Method from C")

class D(B, C):  # Multiple inheritance
    pass

d = D()
d.show()  


# Python uses the C3 Linearization algorithm to resolve the Diamond Problem.
# This ensures that each class appears only once in the MRO and avoids ambiguity.
# You can check the MRO using D.mro().
```

```python
# 14. Write a class method that keeps track of the number of instances created from a class.

# A class method can be used to track the number of instances created from a class:


class Counter:
    instance_count = 0 

    def __init__(self):
        Counter.instance_count += 1

    @classmethod
    def get_instance_count(cls):
        return cls.instance_count

# Example usage
obj1 = Counter()
obj2 = Counter()
print(Counter.get_instance_count())  # Output: 2
```

```python
# 15. Implement a static method in a class that checks if a given year is a leap year.

# A static method can be used when you don't need access to instance or class attributes.
# Here’s how you can implement a leap year checker:

class Calendar:
    @staticmethod
    def is_leap_year(year):
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        return False

# Example usage
print(Calendar.is_leap_year(2024)) 
print(Calendar.is_leap_year(2023)) 
```
